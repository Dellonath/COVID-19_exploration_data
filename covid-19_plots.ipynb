```python
dates = list(pd.date_range('20191231', '20200713').strftime("%Y-%m-%d")) # generating list with 197 date
def total_cases_sum(covid):
    newdata = {}
    for i in dates:
        newdata[i] = covid[covid['date'] == i]['total_cases'].sum()
    return newdata

def total_deaths_sum(covid):
    newdata = {}
    for i in dates:
        newdata[i] = covid[covid['date'] == i]['total_deaths'].sum()
    return newdata
```

```python
# sum of total_cases by continent
south_america_total_cases = total_cases_sum(covid[covid['continent'] == 'South America'])
north_america_total_cases = total_cases_sum(covid[covid['continent'] == 'North America'])
asia_total_cases = total_cases_sum(covid[covid['continent'] == 'Asia'])
africa_total_cases = total_cases_sum(covid[covid['continent'] == 'Africa'])
europe_total_cases = total_cases_sum(covid[covid['continent'] == 'Europe'])
oceania_total_cases = total_cases_sum(covid[covid['continent'] == 'Oceania'])

# sum of total_deaths by continent
south_america_total_deaths = total_deaths_sum(covid[covid['continent'] == 'South America'])
north_america_total_deaths = total_deaths_sum(covid[covid['continent'] == 'North America'])
asia_total_deaths = total_deaths_sum(covid[covid['continent'] == 'Asia'])
africa_total_deaths = total_deaths_sum(covid[covid['continent'] == 'Africa'])
europe_total_deaths = total_deaths_sum(covid[covid['continent'] == 'Europe'])
oceania_total_deaths = total_deaths_sum(covid[covid['continent'] == 'Oceania'])
```

# Comparative plot between continents

```python
plt.figure(figsize=(15,7))
sns.lineplot(np.arange(0, len(south_america_total_cases)), list(south_america_total_cases.values())).set(
                                                                                             xlabel = 'days', 
                                                                                             ylim = (0, 5000000), 
                                                                                             xlim=(0, 200))
sns.lineplot(np.arange(0, len(north_america_total_cases)), list(dict(north_america_total_cases).values()))
sns.lineplot(np.arange(0, len(asia_total_cases)), list(asia_total_cases.values()))
sns.lineplot(np.arange(0, len(africa_total_cases)), list(africa_total_cases.values()))
sns.lineplot(np.arange(0, len(europe_total_cases)), list(europe_total_cases.values()))
sns.lineplot(np.arange(0, len(oceania_total_cases)), list(oceania_total_cases.values()))
plt.legend(labels=['South America', 'North America', 'Asia', 'Africa', 'Europe', 'Oceania'])
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(np.arange(0, len(south_america_total_deaths)), list(south_america_total_deaths.values())).set(
                                                                                             xlabel = 'days', 
                                                                                             ylim = (0,220000), 
                                                                                             xlim=(0, 200))
sns.lineplot(np.arange(0, len(north_america_total_deaths)), list(dict(north_america_total_deaths).values()))
sns.lineplot(np.arange(0, len(asia_total_deaths)), list(asia_total_deaths.values()))
sns.lineplot(np.arange(0, len(africa_total_deaths)), list(africa_total_deaths.values()))
sns.lineplot(np.arange(0, len(europe_total_deaths)), list(europe_total_deaths.values()))
sns.lineplot(np.arange(0, len(oceania_total_deaths)), list(oceania_total_deaths.values()))
plt.legend(labels=['South America', 'North America', 'Asia', 'Africa', 'Europe', 'Oceania'])
```
